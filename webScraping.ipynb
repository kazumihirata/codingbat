```python
import requests
import lxml
from bs4 import BeautifulSoup
import re
import os

# -------------------------------------------------
# Settings
# -------------------------------------------------
base_url = 'http://codingbat.com'
email = 'email@example.com'
password = 'password'

# -------------------------------------------------
# Login
# -------------------------------------------------
client = requests.Session()

HOMEPAGE_URL = base_url
LOGIN_URL = base_url + '/login'

html = client.get(HOMEPAGE_URL).content
soup = BeautifulSoup(html, 'html.parser')

login_information = {
    'uname':email,
    'pw':password,
    'fromurl':'/python'
}

client.post(LOGIN_URL, data=login_information)

# -------------------------------------------------
# Python page
# -------------------------------------------------
target_url = base_url + '/python'
r = client.get(target_url)
soup = BeautifulSoup(r.text, 'html.parser')
soup_a = soup.find_all('a', href=re.compile('/python/.*'))

for i in soup_a :
    # -------------------------------------------------
    # Question List
    # -------------------------------------------------
    dirname = os.path.basename(i['href'])
    target_url = base_url + i['href']
    r = client.get(target_url)
    soup = BeautifulSoup(r.text, 'html.parser')
    soup_a = soup.find_all('a', href=re.compile('/prob/.*'))

    for j in soup_a :
        filename = j.text + '.py'
        # -------------------------------------------------
        # Question Detail
        # -------------------------------------------------
        target_url = base_url + j['href']
        r = client.get(target_url)
        soup = BeautifulSoup(r.text, 'html.parser')
        soup_question = soup.find_all('p', class_='max2')
        soup_answer = soup.find_all('div', id='ace_div')
        print(dirname + "/" + filename)
        print("--------")

        print(soup_question[0].get_text())
        print("--------")
        print(soup_answer[0].get_text())
        print("-----------------------")
```

```python
import glob
import os
path = r"xxxxxxxxxx\codingbat\*.py"
basePath = r"xxxxxxxxxx\codingbat\\"

for f in glob.glob(path) :
    beforeFilename = os.path.basename(f)
    afterFilename = beforeFilename.replace('_', '\\', 1)
    afterDirname = os.path.dirname(afterFilename)

    if not os.path.exists(basePath + afterDirname) :
        os.mkdir(basePath + afterDirname)
    os.rename(basePath + beforeFilename, basePath + afterFilename)
    
```
